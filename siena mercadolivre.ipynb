```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#dados coletados pelo script de webscraping
data = pd.read_csv('carros.csv')
data.info()
data.describe()
```

```python
preco = data.iloc[:,0]
ano = data.iloc[:,1]
kms = data.iloc[:,2]
plt.scatter(kms,preco, color='red')
plt.xlabel('KMs')
plt.ylabel('Preço R$')
#dados dispersos
```

```python
plt.scatter(ano, kms, color='blue')
plt.xlabel('Ano')
plt.ylabel('KMs')
#igualmente dispersos
```
